# Custom iterators x List comprehension i Python

## 1. Iterable - Hvad er det?

```python
family = ['Nicklas', 'Claus', 'Amalie']  # dette er en iterable

print('Her f√•r jeg 2 forskellige iteratorer: \n')

for name in family:
    if name == 'Claus':
        print(name)
        break

print('\n')
for name in family:
    print(name)


iterator = family.__iter__()
print('\n')
print('Her bruger jeg den samme iterator 2 gange: \n')
for name in iterator:
    if name == 'Claus':
        print(name)
        break

for name in iterator:
    print(name)
```
